```python
import matplotlib.image as mpimg
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from PIL import Image

def power_spectrums(image_path):
    # Read and preprocess the image (ensure it's square and grayscale)
    image = mpimg.imread(image_path)
    if image.ndim == 3:
        image = np.mean(image, axis=2)  # Convert to grayscale if RGB
    
    # Crop to square (center crop)
    min_dim = min(image.shape[0], image.shape[1])
    image = image[:min_dim, :min_dim]
    npix = image.shape[0]  # Now npix is consistent for x and y

    # Display the preprocessed image
    plt.figure(figsize=(5, 5))
    plt.imshow(image, cmap='gray')
    plt.title("Preprocessed Image (Square)")
    plt.axis('off')
    plt.show()

    # Compute FFT and power spectrum
    fourier_image = np.fft.fftn(image)
    fourier_amplitudes = np.abs(fourier_image)**2
    fourier_2d = np.fft.fftshift(fourier_amplitudes)

    # Compute radial frequencies
    kfreq = np.fft.fftfreq(npix) * npix
    kfreq2D = np.meshgrid(kfreq, kfreq)
    knrm = np.sqrt(kfreq2D[0]**2 + kfreq2D[1]**2)

    # Flatten arrays for radial averaging
    knrm_flat = knrm.flatten()
    fourier_flat = fourier_amplitudes.flatten()

    # Debug: Check shapes
    print(f"Shape of knrm_flat: {knrm_flat.shape}")
    print(f"Shape of fourier_flat: {fourier_flat.shape}")
    assert len(knrm_flat) == len(fourier_flat), "Mismatch in flattened array lengths!"

    # Bin the power spectrum
    kbins = np.arange(0.5, npix//2 + 1, 1.)
    kvals = 0.5 * (kbins[1:] + kbins[:-1])
    Abins, _, _ = stats.binned_statistic(knrm_flat, fourier_flat,
                                       statistic="mean",
                                       bins=kbins)
    Abins *= np.pi * (kbins[1:]**2 - kbins[:-1]**2)

    # Plot 1D power spectrum
    plt.figure(figsize=(8, 6))
    plt.loglog(kvals, Abins)
    plt.xlabel("$k$")
    plt.ylabel("$P(k)$")
    plt.title("Radially Averaged Power Spectrum (1D)")
    plt.grid(True, which="both")
    plt.tight_layout()
    plt.show()

    # Plot 2D power spectrum
    plt.figure(figsize=(6, 6))
    plt.imshow(np.log1p(fourier_2d), cmap='inferno')
    plt.colorbar(label="log(Power)")
    plt.title("2D Power Spectrum")
    plt.axis('off')
    plt.show()
```
